```python
input_data = ['190: 10 19','83: 17 5','156: 78 2']
totalCalibrationResult = 0

def convertStringToNumberList(stringToProcess):
    result = stringToProcess.split(': ')
    second_result = result[1].split()
    result.pop(1)
    final_result = result + second_result
    for index in range(len(final_result)):
        final_result[index] = int(final_result[index])
    return final_result
    
def operatorChecker(numberListToCheck):
    answer = numberListToCheck[0]
    if numberListToCheck[1] + numberListToCheck[2] == answer:
        return True
    elif numberListToCheck[1] * numberListToCheck[2] == answer:
        return True
    return False

    
for line in input_data:
    currentNumberList = convertStringToNumberList(line)
    print(currentNumberList)
    if operatorChecker(currentNumberList) is True:
        totalCalibrationResult += currentNumberList[0]
    print(totalCalibrationResult)
```
